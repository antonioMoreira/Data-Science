## Wine dataset

Usando apenas duas classes do conjunto de dados:
> $w_1$ → _Class 1_, com 59 amostras<br>
> $w_2$ → _Class 2_, com 71 amostras<br>

A dimensionalidade representa o número de _features_ (variáveis aleatórias, estatísticamente falando) do conjunto de dados, neste caso 13 no total, contudo, serão usadas apenas as 3 primeiras:
> 1. _Alcohol_ <br>
> 2. _Malic Acid_ <br>
> 3. _Ash_ <br>

Deste modo:
$$ P(w_1) = \frac{59}{130} \approx 0.454$$

$$ P(w_2) = \frac{71}{130} \approx 0.546$$


Seja *x* um conjunto de _features_ desconhecidas: $ \vec{x} = [x_1,x_2,x_3]$

```python
from sklearn.datasets import load_wine

data = load_wine()
#print("Features: ", data.feature_names)

# Slicing dataset
X = data.data[:130,:3]
y = data.target[:130]

# Unknown feature vector 
x = np.array([13.243, 2.253, 2.22])
```

Sabemos (_ou estimaremos_) a _class-conditional_ **pdf** (_probability desity function_), também conhecida como **função de máxima verossimilhança de $w_i$ com respeito à _x_**:
$$p(\vec{x}|w_i), i=1,2$$

Plotando todas as _class-conditionals_ **pdf's**, onde os gráficos em azul representam a variável alatória $X_i,i=1,2,3$ com respeito à _Classe 1_ e em vermelho à _Classe 2_.

```python
(f, axes) = plt.subplots(2, 2, figsize=(15, 15))

sns.distplot(X[:59,0], color="b", ax=axes[0,0], kde_kws={"label":"Class 1"})
sns.distplot(X[59:,0], color="r", ax=axes[0,0], kde_kws={"label":"Class 2"})
axes[0,0].set_title("Alcohol", fontdict={'fontsize':18})

sns.distplot(X[:59,1], color="b", ax=axes[0,1], kde_kws={"label":"Class 1"})
sns.distplot(X[59:,1], color="r", ax=axes[0,1], kde_kws={"label":"Class 2"})
axes[0,1].set_title("Malic Acid", fontdict={'fontsize':18})

sns.distplot(X[:59,2], color="b", ax=axes[1,0], kde_kws={"label":"Class 1"})
sns.distplot(X[59:,2], color="r", ax=axes[1,0], kde_kws={"label":"Class 2"})
axes[1,0].set_title("Ash", fontdict={'fontsize':18})

axes[1,1].set_axis_off()

# -----------------

x1_w1 = axes[0,0].get_contains()
x1_w1
```

Nosso **objetivo** é: _"Dado um conjunto de características determinar as probabilidades de pertencer à classe $w_i$ e classificá-lo de acordo com a maior probabilidade encontrada"_, isto é, as probabilidades de uma evidência _x_ pertencer às classes $w_1$ e $w_2$ são, respectivamente, *0.3* e *0.7* , deste modo, como *0.3 < 0.7*, *x* é classificado como $w_1$.

---

→ _Probabilidade a posteriori_: prob. do parâmetro $\theta$ dada a evidência _X_: 
$$P(\theta|X) = \frac{P(X|\theta)P(\theta)}{P(X)}$$ 

→ _Maximum a posteriori estimation (**MAP**)_:
$$\theta \mapsto f(\theta|x) = \frac{f(x|\theta)g(\theta)}{\int_\Theta f(x|\mathcal{v})g(\mathcal{v})d\mathcal{v}}$$

$$\hat{\theta}_{MAP}(x) = \underset{\theta}{arg max} f(\theta|x)$$

---

Voltando ao nosso caso, pela regra de _Bayes_: 
$$P(w_i|\vec{x}) = \frac{p(\vec{x}|w_i)P(w_i)}{p(\vec{x})}$$

Onde a **pdf** de $\vec{x}$ é:
$$ p(\vec{x}) = \displaystyle\sum_{i=1}^{2} p(\vec{x}|w_i)P(w_i)$$
